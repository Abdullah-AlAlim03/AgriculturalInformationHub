```python
from abc import ABC, abstractmethod
class triangle(ABC):
  def __init__(self, base, height):
    self.base = base
    self.height = height

  @abstractmethod
  def area(self):
    pass

class calculation(triangle):
  def area(self):
    self.area = 0.5 * self.base * self.height
    print(self.area)

c1=calculation(10, 12)
c1.area()
```
